```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
get_link = 'https://books.toscrape.com/catalogue/page-1.html'
res = requests.get(get_link)
soup = BeautifulSoup(res.text, 'html.parser')
list_books = soup.find('ol',{'class':'row'})
books_item = list_books.find_all('li')

page_num = soup.find('li',{'class':'current'}).text
text = page_num.strip()
total_pages = text.split(' of ')[1]
total_pages = int(total_pages)


with open('all_books.csv', mode = 'w', encoding = 'utf-8') as fd:
    fd.write('Books, Price, Status, Rate \n')

    for i in range (total_pages):
        link = f'https://books.toscrape.com/catalogue/page-{i+1}.html'
        res_2 = requests.get(link)
        soup_2 = BeautifulSoup(res_2.text, 'html.parser')
        list_books_2 = soup_2.find('ol',{'class':'row'})
        books_item_2 = list_books_2.find_all('li')
    
        for books in range (len(books_item_2)):
            books_name = list_books_2.find_all('h3')[books].text
            price = list_books_2.find_all('p',{'class':'price_color'})[books].text
            status = list_books_2.find_all('p',{'class':'instock availability'})[books].text.split('\n')[3].strip()
            rate = list_books.find_all('article',{'class':'product_pod'})[books].find('p').attrs['class']
            rate_stars = ' - '.join(rate)
            fd.write(f'\"{books_name}\", \"{price}\", \"{status}\", {rate_stars} \n')
    
```
